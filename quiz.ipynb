```python
import json
import random

# Sample question bank
question_bank = {
    "electromagnetism": {
        "easy": [
            {"question": "What is the unit of electric current?", "options": ["Ampere", "Volt", "Ohm", "Watt"], "answer": "Ampere"},
            {"question": "What does a capacitor store?", "options": ["Charge", "Voltage", "Current", "Energy"], "answer": "Charge"},
        ],
        "medium": [
            {"question": "What is the formula for calculating electric power?", "options": ["P=IV", "P=I^2R", "P=V^2/R", "All of the above"], "answer": "All of the above"},
            {"question": "What is the principle of electromagnetic induction?", "options": ["Faraday's Law", "Ohm's Law", "Newton's Law", "Coulomb's Law"], "answer": "Faraday's Law"},
        ],
        "difficult": [
            {"question": "What is the relationship between electric field and potential?", "options": ["E = -dV/dx", "E = V/dx", "E = V^2", "E = IV"], "answer": "E = -dV/dx"},
            {"question": "What is the magnetic field inside a solenoid?", "options": ["B = μ₀(nI)", "B = μ₀I", "B = I/R", "B = V/I"], "answer": "B = μ₀(nI)"},
        ],
    },
    "classical_mechanics": {
        "easy": [
            {"question": "What is the force that opposes motion?", "options": ["Friction", "Gravity", "Magnetism", "Inertia"], "answer": "Friction"},
            {"question": "What is the formula for speed?", "options": ["Distance/Time", "Time/Distance", "Distance*Time", "Distance+Time"], "answer": "Distance/Time"},
        ],
        "medium": [
            {"question": "What is Newton's second law?", "options": ["F=ma", "E=mc^2", "F=mv", "F=ma^2"], "answer": "F=ma"},
            {"question": "What is the principle of conservation of momentum?", "options": ["Total momentum before = Total momentum after", "Total energy before = Total energy after", "Total force before = Total force after", "None of the above"], "answer": "Total momentum before = Total momentum after"},
        ],
        "difficult": [
            {"question": "What is the moment of inertia?", "options": ["I = Σmr²", "I = mv²", "I = mgh", "I = Fd"], "answer": "I = Σmr²"},
            {"question": "What is the work-energy theorem?", "options": ["W = ΔKE", "W = Fd", "W = mgh", "W = P*t"], "answer": "W = ΔKE"},
        ],
    },
    # Add more topics as needed
}

# Function to greet the user and collect information
def greet_and_collect_info():
    # Greet the user in English
    print("Hello! Welcome to the Science Chatbot.")
    
    # Language options
    languages = ["English", "Spanish", "French", "German", "Chinese", "Japanese", "Russian", "Italian", "Portuguese", "Hindi"]
    print("Please choose a language:")
    for i, lang in enumerate(languages, 1):
        print(f"{i}. {lang}")
    
    lang_choice = int(input("Enter the number corresponding to your choice: ")) - 1
    chosen_language = languages[lang_choice]
    
    # Greet the user in the chosen language (for simplicity, only English greeting is shown)
    print(f"Welcome! You have chosen {chosen_language}.")
    
    # Ask for user details
    name = input("What is your name? ")
    age = int(input("How old are you? "))
    
    # Save user information to a JSON file
    user_info = {
        "name": name,
        "age": age,
        "language": chosen_language
    }
    
    with open('user_info.json', 'w') as json_file:
        json.dump(user_info, json_file)
    
    print("Your information has been saved in 'user_info.json'.")
    
    # Provide options for functionalities
    print("What would you like to do?")
    print("1. Take a Quiz")
    print("2. Use the Calculator")
    
    choice = int(input("Enter the number corresponding to your choice: "))
    
    if choice == 1:
        take_quiz(age)
    elif choice == 2:
        print("Calculator functionality is not implemented yet.")
    else:
        print("Invalid choice. Please try again.")

# Function to conduct a quiz based on age
def take_quiz(age):
    if age <= 5:
        difficulty = "easy"
    elif 5 < age <= 18:
        difficulty = "medium"
    else:
        difficulty = "difficult"
    
    print(f"You will be answering {difficulty} level questions.")
    num_questions = int(input("How many questions would you like to answer? (5 or 10): "))
    
    if num_questions not in [5, 10]:
        print("Invalid number of questions. Please choose either 5 or 10.")
        return
    
    selected_questions = []
    for topic in question_bank:
        selected_questions.extend(random.sample(question_bank[topic][difficulty], min(num_questions, len(question_bank[topic][difficulty]))))
    
    score = 0
    feedback = {topic: 0 for topic in question_bank}
    
    for question in selected_questions:
        print(question["question"])
        for i, option in enumerate(question["options"], 1):
            print(f"{i}. {option}")
        
        answer = int(input("Enter the number of your answer: ")) - 1
        if question["options"][answer] == question["answer"]:
            score += 1
            feedback[topic] += 1
    
    print(f"You scored {score} out of {num_questions}.")
    for topic, correct in feedback.items():
        if correct > 0:
            print(f"You answered {correct} questions correctly in {topic}.")
        else:
            print(f"You need to work on {topic} more.")

# Call the function to start the interaction
greet_and_collect_info()
```

```python
import json

# Language greeting dictionary
greetings = {
    "English": "Hello",
    "Spanish": "Hola",
    "French": "Bonjour",
    "German": "Hallo",
    "Chinese": "你好",
    "Japanese": "こんにちは",
    "Hindi": "नमस्ते",
    "Arabic": "مرحبا",
    "Russian": "Привет",
    "Italian": "Ciao"
}

# Question Bank
question_bank = {
    "Electromagnetism": {
        "Easy": [
            {"question": "What is the unit of electric current?", "options": ["Ampere", "Volt", "Ohm", "Watt"], "answer": "Ampere"},
            {"question": "Which particle carries a negative charge?", "options": ["Electron", "Proton", "Neutron", "Photon"], "answer": "Electron"}
        ],
        "Medium": [
            {"question": "What is the speed of light in vacuum?", "options": ["3x10^8 m/s", "2x10^8 m/s", "1x10^8 m/s", "5x10^8 m/s"], "answer": "3x10^8 m/s"}
        ],
        "Difficult": [
            {"question": "What is the formula for the Lorentz force?", "options": ["F = q(E + v×B)", "F = qE", "F = v×B", "F = E/B"], "answer": "F = q(E + v×B)"}
        ]
    },
    "Classical Mechanics": {
        "Easy": [
            {"question": "What is the unit of force?", "options": ["Newton", "Joule", "Pascal", "Watt"], "answer": "Newton"}
        ],
        "Medium": [
            {"question": "What is the second law of motion by Newton?", "options": ["F=ma", "E=mc^2", "V=IR", "P=W/t"], "answer": "F=ma"}
        ],
        "Difficult": [
            {"question": "What is the formula for kinetic energy?", "options": ["KE=1/2mv^2", "KE=mv^2", "KE=1/2mv", "KE=mv"], "answer": "KE=1/2mv^2"}
        ]
    },
    "Quantum Mechanics": {
        "Easy": [
            {"question": "What does a photon represent?", "options": ["A quantum of light", "A particle of mass", "A type of charge", "A wave"], "answer": "A quantum of light"}
        ],
        "Medium": [
            {"question": "Who is known as the father of quantum mechanics?", "options": ["Max Planck", "Einstein", "Newton", "Bohr"], "answer": "Max Planck"}
        ],
        "Difficult": [
            {"question": "What is the formula for the energy of a photon?", "options": ["E=hf", "E=mc^2", "E=VIt", "E=1/2mv^2"], "answer": "E=hf"}
        ]
    }
}


# Save user data to JSON
def save_user_data(user_data):
    try:
        with open("user_data.json", "w") as file:
            json.dump(user_data, file, indent=4)
    except Exception as e:
        print("Error saving user data:", e)


# Quiz Function
def run_quiz(age):
    print("\nChoose a topic:")
    topics = list(question_bank.keys())
    for idx, topic in enumerate(topics, start=1):
        print(f"{idx}. {topic}")
    
    topic_choice = int(input("Enter your choice (1/2/3): ")) - 1
    topic = topics[topic_choice]
    
    difficulty = "Easy" if age <= 5 else "Medium" if age <= 18 else "Difficult"
    questions = question_bank[topic][difficulty]
    
    num_questions = int(input("\nHow many questions do you want? (5/10): "))
    score = 0
    selected_questions = questions[:num_questions]
    
    for idx, q in enumerate(selected_questions, start=1):
        print(f"\nQuestion {idx}: {q['question']}")
        for i, option in enumerate(q["options"], start=1):
            print(f"{i}. {option}")
        answer = int(input("Enter your answer (1/2/3/4): "))
        if q["options"][answer - 1] == q["answer"]:
            score += 1
    
    print(f"\nYour score: {score}/{num_questions}")
    
    feedback = f"You answered {score} questions correctly in {topic}."
    if score < num_questions // 2:
        feedback += f" You might want to study more on {topic}."
    else:
        feedback += f" Great job on {topic}!"
    
    print(feedback)


# Main Function
def main():
    print("Welcome to the Science Chatbot!")
    
    # Language selection
    print("\nChoose your language:")
    languages = list(greetings.keys())
    for idx, lang in enumerate(languages, start=1):
        print(f"{idx}. {lang}")
    
    lang_choice = int(input("Enter your choice (1-10): ")) - 1
    selected_language = languages[lang_choice]
    print(f"\n{greetings[selected_language]}!")
    
    # User details
    name = input("\nWhat is your name? ")
    age = int(input("What is your age? "))
    user_data = {"name": name, "age": age}
    save_user_data(user_data)
    
    # Main menu
    while True:
        print("\nWhat would you like to do?")
        print("1. Take a Quiz")
        print("2. Use Calculator")
        print("3. Exit")
        choice = int(input("Enter your choice (1/2/3): "))
        
        if choice == 1:
            run_quiz(age)
        elif choice == 2:
            print("\nCalculator is under development.")
        elif choice == 3:
            print("\nGoodbye!")
            break
        else:
            print("\nInvalid choice, please try again.")


if __name__ == "__main__":
    main()
```
